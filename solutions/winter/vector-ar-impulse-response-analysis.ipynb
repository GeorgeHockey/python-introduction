### Exercise 71
Plot the impulse responses from both a first order model and
the model selected in the order selection exercise.

```python
import pandas as pd
import statsmodels.tsa.api as tsa
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
plt.rc("figure", figsize=(16,12))

data = pd.read_hdf("./data/var-data.h5", "var_data")
var_data = data[["spread","gs1","deflg"]]
mod = tsa.VAR(var_data)
res = mod.fit(2, ic=None)

irfs = res.irf(12)
fig = irfs.plot()
```

```python
_ = irfs.plot(True)
```
